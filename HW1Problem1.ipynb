```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
def f(x, y):
    return (5 * x ** 2)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
```

```
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.contour3D(X, Y, Z, 100, cmap='Blues_r')
ax.scatter3D(0, 0, 0, c='r')
ax.view_init(30, 80)
plt.show()
```
